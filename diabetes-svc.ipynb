**Data preprocessing**

```python
x = df.drop(columns='Outcome')
y = df['Outcome'] 
```

#  **Model** 

**model usign pipline:**

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test  = train_test_split(x,y,random_state=42)
pipe = Pipeline([('scaler', StandardScaler()), ('svc', SVC())])
pipe.fit(X_train, y_train).score(X_test, y_test)
```

```python
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler 
from sklearn.metrics import accuracy_score,confusion_matrix

sc =StandardScaler()
x_scaled = sc.fit_transform(x)

X_train,X_test,y_train,y_test  = train_test_split(x_scaled,y,random_state=42)
 

# Creating the SVC model with the specified kernel
model = SVC(kernel='rbf', gamma=0.04)
model.fit(X_train,y_train)
ypred = model.predict(X_test)
score = accuracy_score(y_test,ypred)
print('score:')
print(score)
print('matrix:')
matrix = confusion_matrix(y_test,ypred)
print(matrix)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, confusion_matrix

# Création de l'instance du modèle
dt = DecisionTreeClassifier()

# Entraînement du modèle
dt.fit(X_train, y_train)

# Prédiction sur les données de test
prediction = dt.predict(X_test)

# Calcul du score d'exactitude
scoreDecisionTreeClassifier = accuracy_score(y_test, prediction)
print('Score:')
print(scoreDecisionTreeClassifier)

# Calcul de la matrice de confusion
conf_matrix = confusion_matrix(y_test, prediction)
print('Matrice de confusion:')
print(conf_matrix)
```

```python
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt
columns = df.columns
plt.figure(figsize=(20, 10))  
plot_tree(dt, filled=True, feature_names=df.drop(columns='Outcome').columns, class_names=columns)  
plt.show()
```
